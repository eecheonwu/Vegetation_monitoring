```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
# Sample data (replace with your actual values)
model_predictions = [1.00, 1.00, 1.00, 0.84, 1.00, 1.00, 1.00, 0.97, 0.52, 0.21, 0.59, 0.09, 0.16, 0.20, 0.33, 0.11, 0.05]
ground_truth = [1.00, 0.96, 0.98, 0.78, 0.91, 0.96, 0.93, 0.95, 0.52, 0.15, 0.53, 0.07, 0.10, 0.17, 0.35, 0.12, 0.00]
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(model_predictions, ground_truth, s=40)  # Adjust marker size as needed

# Add the diagonal line for perfect prediction
plt.plot([0, 1], [0, 1], color='black', linestyle='--', label='Perfect Prediction')

# Customize the plot
plt.xlabel("Model Predictions", fontsize=14)
plt.ylabel("Ground Truth Values", fontsize=14)
plt.title("Scatter Plot of Model Predictions vs. Ground Truth", fontsize=16)
plt.xlim(0, 1)  # Set appropriate x-axis limits
plt.ylim(0, 1)  # Set appropriate y-axis limits
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
model_predictions = np.array(model_predictions)  # Convert lists to NumPy arrays
ground_truth = np.array(ground_truth) 
```

```python
MAE = np.mean(np.abs(model_predictions - ground_truth))
MSE = np.mean((model_predictions - ground_truth)**2)
RMSE = np.sqrt(MSE)
SS_res = np.sum((model_predictions - ground_truth)**2)
SS_tot = np.sum((ground_truth - np.mean(ground_truth))**2)
R_squared = 1 - (SS_res / SS_tot)
```

```python
print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
print("R-squared: ", R_squared)
```

```python
errors = model_predictions - ground_truth
```

```python
# Create the error distribution plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.hist(errors, bins=15, edgecolor='black')  # Adjust bin number as needed

# Customize the plot
plt.xlabel("Prediction Errors", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.title("Distribution of Prediction Errors", fontsize=16)
plt.xlim(-0.4, 0.4)  # Adjust x-axis limits as needed
plt.grid(True)

# Show the plot
plt.show()
```
